```r
# Load libraries
library(tidyverse)
install.packages("tidyverse")

set.seed(2025)

# Number of rows
n <- 100000

# Generate synthetic data
customer_id <- 1:n

# Randomly assign groups (50% control, 50% treatment)
group <- sample(c("A", "B"), n, replace = TRUE)

# Demographics
age <- round(rnorm(n, mean = 40, sd = 10))
income <- round(rnorm(n, mean = 80000, sd = 20000))
regions <- c("North", "South", "East", "West")
region <- sample(regions, n, replace = TRUE)

# Pre-offer spend (3 months)
spend_before <- round(rnorm(n, mean = 1500, sd = 400))

# Earned points based on spend (1 point per $1)
points_earned <- spend_before * runif(n, 0.8, 1.2)

# Post-offer spend depends on group
# Control (A): slight natural growth
# Treatment (B): increased spend due to offer
spend_after <- ifelse(group == "A",
                      spend_before * runif(n, 0.95, 1.15),
                      spend_before * runif(n, 1.05, 1.35))

# Redeemed points: depends on group and engagement
redeemed_points <- ifelse(group == "A",
                          pmax(0, rpois(n, lambda = points_earned * 0.3 / 100)),
                          pmax(0, rpois(n, lambda = points_earned * 0.5 / 100)))

# Engagement score: scaled 0-100
# Treatment group tends to have higher scores
engagement_score <- pmin(100,
                         round(ifelse(group == "A",
                                      rnorm(n, mean = 50, sd = 15),
                                      rnorm(n, mean = 60, sd = 15))))

# Combine into dataframe
data <- tibble(
  customer_id,
  group,
  age,
  income,
  region,
  spend_before_offer = spend_before,
  spend_after_offer = round(spend_after, 2),
  points_earned = round(points_earned, 0),
  redeemed_points = redeemed_points,
  engagement_score = engagement_score
)

# Save dataset
write_csv(data, "marketing_ab_test_large.csv")

# Quick check
glimpse(data)
```

```r
# Step 1: Load Libraries
library(tidyverse)

# Step 2: Load Data
data <- read_csv("marketing_ab_test_large.csv")

# Inspect data
glimpse(data)
summary(data)

# Step 3: Data Cleaning
# Handle missing values
data <- data %>%
  mutate(across(where(is.numeric), ~ ifelse(is.na(.), median(., na.rm=TRUE), .)))

# Convert categorical columns
data <- data %>%
  mutate(group = as.factor(group),
         region = as.factor(region))

# Remove outliers (1% tails for spend_after_offer)
q <- quantile(data$spend_after_offer, probs = c(0.01, 0.99))
data <- data %>%
  filter(spend_after_offer >= q[1] & spend_after_offer <= q[2])

# Feature Engineering
data <- data %>%
  mutate(
    redeemed_flag = ifelse(redeemed_points > 0, 1, 0),
    spend_change = spend_after_offer - spend_before_offer
  )

# Save cleaned data
write_csv(data, "marketing_ab_test_clean.csv")

# Step 4: EDA
# Group Summary
data %>%
  group_by(group) %>%
  summarise(
    avg_spend_before = mean(spend_before_offer),
    avg_spend_after = mean(spend_after_offer),
    avg_points_redeemed = mean(redeemed_points),
    avg_engagement = mean(engagement_score),
    n = n()
  )

# Distribution of post-offer spend
ggplot(data, aes(x=spend_after_offer, fill=group)) +
  geom_histogram(position="dodge", bins=40) +
  labs(title="Post-Offer Spend by Group", x="Spend After Offer", y="Count") +
  theme_minimal()

# Engagement Score Distribution
ggplot(data, aes(x=engagement_score, fill=group)) +
  geom_density(alpha=0.5) +
  labs(title="Engagement Score Distribution by Group") +
  theme_minimal()

# Earned vs Redeemed Points
data %>%
  ggplot(aes(x=points_earned, y=redeemed_points, color=group)) +
  geom_point(alpha=0.3) +
  geom_smooth(method="lm", se=FALSE) +
  labs(title="Earned vs Redeemed Points", x="Points Earned", y="Redeemed Points") +
  theme_minimal()

# Step 5: Hypothesis Testing
# Balance check
t.test(spend_before_offer ~ group, data=data)

# Post-offer spend impact
t.test(spend_after_offer ~ group, data=data)

# Lift Calculation
mean_A <- mean(filter(data, group=="A")$spend_after_offer)
mean_B <- mean(filter(data, group=="B")$spend_after_offer)
lift <- (mean_B - mean_A) / mean_A * 100
paste("Lift due to offer:", round(lift, 2), "%")

# Engagement Score Impact
t.test(engagement_score ~ group, data=data)

# Redemption Rate
data %>%
  group_by(group) %>%
  summarise(redemption_rate = mean(redeemed_flag))

# Step 6: Visualization for Stakeholders
# Boxplot
ggplot(data, aes(x=group, y=spend_after_offer, fill=group)) +
  geom_boxplot() +
  labs(title="Impact of Offer on Post-Offer Spend", x="Group", y="Spend After Offer") +
  theme_minimal()

# Redemption Rate by Group
data %>%
  group_by(group) %>%
  summarise(redemption_rate = mean(redeemed_flag)) %>%
  ggplot(aes(x=group, y=redemption_rate, fill=group)) +
  geom_col() +
  scale_y_continuous(labels=scales::percent) +
  labs(title="Redemption Rate by Group", x="Group", y="Redemption Rate") +
  theme_minimal()

# Step 7: Advanced Regression Analysis
model <- lm(spend_after_offer ~ group + age + income + spend_before_offer, data=data)
summary(model)
```
